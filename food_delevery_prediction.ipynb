```python
avr=[]
au=(np.unique(df["Average_Cost"]))
for i in np.arange(len(df["Average_Cost"])):
    if (df["Average_Cost"][i])==au[0]:
         avr=np.append(avr,0)
    elif len(df["Average_Cost"][i])==len(au[1]):
        avr=np.append(avr,(int(df["Average_Cost"][i][1:2]+df["Average_Cost"][i][3:6])))
    elif len(df["Average_Cost"][i])==len(au[10]):
        avr=np.append(avr,int(df["Average_Cost"][i][1:4])) 
    elif len (df["Average_Cost"][i])==len(au[15]):
        avr=np.append(avr,int(df["Average_Cost"][i][1:3])) 
    else :
        continue
```

```python
mo=[]
mu=(np.unique(df["Minimum_Order"]))
for i in np.arange(len(df["Minimum_Order"])):
    if len(df["Minimum_Order"][i])==len(mu[0]):
        mo=np.append(mo,int(df["Minimum_Order"][i][1:2]))
    elif len(df["Minimum_Order"][i])==len(mu[1]):
        mo=np.append(mo,int(df["Minimum_Order"][i][1:4])) 
    elif len(df["Minimum_Order"][i])==len(mu[11]):
        mo=np.append(mo,int(df["Minimum_Order"][i][1:3])) 
```

```python
def outliers(ar):
    m=np.mean(ar)
    sd=np.std(ar)
    ud=m+(3*sd)
    ld=m-(3*sd)
    ser=[]
    v=[]
    for i in np.arange(len(ar)):
         if (ar[i]>=ld) & (ar[i]<=ud):
                v=np.append(v,ar[i])
                ser=np.append(ser,i)
    return(DataFrame([ser,v]).T)
```
